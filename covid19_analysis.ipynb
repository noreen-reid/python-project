```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset using your file path
file_path = r"C:\Users\disma\OneDrive\Desktop\covid19-analysis\cumulative-deaths-and-cases-covid-19.csv"
df = pd.read_csv(file_path)

# Preview the dataset
print(f"Shape: {df.shape}")
df.head()
```

```python
df.columns = ['country', 'iso_code', 'date', 'total_deaths', 'total_cases']

```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
countries = ['Kenya', 'United States', 'India']
df_countries = df[df['country'].isin(countries)]
```

```python
df_countries = df_countries.dropna(subset=['total_cases', 'total_deaths'])
```

```python

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_cases', hue='country')
plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_deaths', hue='country')
plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Load the vaccine doses data
vaccine_doses_df = pd.read_csv(r"C:\Users\disma\OneDrive\Desktop\covid19-analysis\covid-vaccine-doses-by-manufacturer.csv")

# Check the first few rows
vaccine_doses_df.head()
```

```python
# Explore columns in the dataset
vaccine_doses_df.columns
```

```python
# Check the unique locations (countries) in the vaccine doses dataset
vaccine_doses_df['Entity'].unique()
```

```python

# Filter for your countries of interest: Kenya, USA, India
countries = ['Kenya', 'United States', 'India']
vaccine_doses_df = vaccine_doses_df[vaccine_doses_df['Entity'].isin(countries)]

# Verify the filter
vaccine_doses_df['Entity'].unique()
```

```python
# Check the column names to identify the date column
vaccine_doses_df.columns
```

```python
# Melt the dataframe to long format, using 'Day' as the date column
vaccine_doses_melted = pd.melt(vaccine_doses_df,
                               id_vars=['Entity', 'Day'],  # Use 'Day' for the date column
                               value_vars=['COVID-19 doses (cumulative) - Manufacturer Moderna',
                                           'COVID-19 doses (cumulative) - Manufacturer Oxford/AstraZeneca',
                                           'COVID-19 doses (cumulative) - Manufacturer Johnson&Johnson',
                                           'COVID-19 doses (cumulative) - Manufacturer Sputnik V',
                                           'COVID-19 doses (cumulative) - Manufacturer Sinovac',
                                           'COVID-19 doses (cumulative) - Manufacturer Sinopharm/Beijing',
                                           'COVID-19 doses (cumulative) - Manufacturer CanSino',
                                           'COVID-19 doses (cumulative) - Manufacturer Novavax',
                                           'COVID-19 doses (cumulative) - Manufacturer Covaxin',
                                           'COVID-19 doses (cumulative) - Manufacturer Medicago',
                                           'COVID-19 doses (cumulative) - Manufacturer Sanofi/GSK',
                                           'COVID-19 doses (cumulative) - Manufacturer SKYCovione',
                                           'COVID-19 doses (cumulative) - Manufacturer Valneva'],
                               var_name='manufacturer', value_name='total_doses')

# Check the reshaped dataframe
vaccine_doses_melted.head()
```

```python
# Plot total vaccine doses administered by manufacturer for each country
plt.figure(figsize=(14, 8))
sns.barplot(data=vaccine_doses_melted, x='Entity', y='total_doses', hue='manufacturer', ci=None)
plt.title('COVID-19 Vaccine Doses Administered by Manufacturer')
plt.xlabel('Country')
plt.ylabel('Total Doses')
plt.legend(title='Manufacturer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python


# Set a color palette for better distinction between manufacturers
palette = sns.color_palette("Set2", len(vaccine_doses_melted['manufacturer'].unique()))

# Plot total vaccine doses administered by manufacturer for each country
plt.figure(figsize=(16, 8))

sns.barplot(data=vaccine_doses_melted, 
            x='Entity', 
            y='total_doses', 
            hue='manufacturer', 
            ci=None, 
            palette=palette)

# Title and labels
plt.title('COVID-19 Vaccine Doses Administered by Manufacturer', fontsize=18)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Total Doses Administered', fontsize=14)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add annotations on top of the bars to display the exact dose numbers
for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height()):,}', 
                       (p.get_x() + p.get_width() / 2., p.get_height()), 
                       ha='center', va='center', 
                       fontsize=12, color='black', 
                       xytext=(0, 9), textcoords='offset points')

# Improve the legend placement
plt.legend(title='Manufacturer', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot the cumulative vaccinations over time for the selected countries
plt.figure(figsize=(12, 6))
sns.lineplot(data=vaccine_doses_melted[vaccine_doses_melted['Entity'].isin(countries)], 
             x='Day', y='total_doses', hue='Entity', marker='o')
plt.title('Cumulative COVID-19 Vaccinations Over Time (Kenya, USA, India)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Vaccinations', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# You would need to load population data or use approximate values for each country
# Example: Population data
population_data = {
    'Kenya': 53771296,
    'United States': 331893745,
    'India': 1393409038
}

# Calculate % vaccinated (you can use a recent snapshot or cumulative value)
vaccine_doses_melted['percent_vaccinated'] = vaccine_doses_melted['total_doses'] / vaccine_doses_melted['Entity'].map(population_data) * 100

# Plot the vaccination rate (percentage) for each country
plt.figure(figsize=(12, 6))
sns.lineplot(data=vaccine_doses_melted[vaccine_doses_melted['Entity'].isin(countries)], 
             x='Day', y='percent_vaccinated', hue='Entity', marker='o')
plt.title('COVID-19 Vaccination Progress (% of Population Vaccinated) Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Vaccination Percentage (%)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
